```python
data_dir = '/path/to/data/'
fiftyWordDat = sio.loadmat(data_dir+'t12.2022.05.03_fiftyWordSet.mat')
sentenceDat = sio.loadmat(data_dir+'t12.2022.04.28_sentences.mat')
```

```python
title = 'Words: average BA6v and BA44 activity'
figPath = 'new_figs/words_timecourse'
plot_trial_timecourse(fiftyWordDat,title=title,saveFig=False,log=True)
```

```python
title = 'Sentences: average BA6v and BA44 activity'
figPath = 'new_figs/sentences_timecourse'
plot_trial_timecourse(sentenceDat,title=title,rel_start=-200,rel_end=500,log=True,saveFig=False)
```

```python
word_zeros, word_ones = calc_trial_mean(fiftyWordDat)
word_delay_6v = extract_channel_data(word_zeros,area_6v)
word_speak_6v = extract_channel_data(word_ones,area_6v)
word_delay_44 = extract_channel_data(word_zeros,area_44)
word_speak_44 = extract_channel_data(word_ones,area_44)
chan_avg_arrays = [np.mean(word_delay_6v,axis=0),
                   np.mean(word_speak_6v,axis=0),
                   np.mean(word_delay_44,axis=0),
                   np.mean(word_speak_44,axis=0)]
```

```python
colors = ['b','g','y','c']
labels = ['delay 6v','speak 6v', 'delay 44', 'speak 44']
title = 'Words: average activity in delay vs. speak conditions'
figPath = 'new_figs/word_channel_hist'
plot_bandpower_histograms(chan_avg_arrays,title=title,labels=labels,colors=colors,figsize=(10, 6),alpha=0.5,saveFig=False)
```

```python
sent_zeros, sent_ones = calc_trial_mean(sentenceDat)
sent_delay_6v = extract_channel_data(sent_zeros,area_6v)
sent_speak_6v = extract_channel_data(sent_ones,area_6v)
sent_delay_44 = extract_channel_data(sent_zeros,area_44)
sent_speak_44 = extract_channel_data(sent_ones,area_44)
sent_avg_arrays = [np.mean(sent_delay_6v,axis=0),
                   np.mean(sent_speak_6v,axis=0),
                   np.mean(sent_delay_44,axis=0),
                   np.mean(sent_speak_44,axis=0)]
```

```python
colors = ['b','g','y','c']
labels = ['delay 6v','speak 6v', 'delay 44', 'speak 44']
title = 'Sentences: average activity in delay vs. speak conditions'
figPath = 'new_figs/sentence_channel_hist'
plot_bandpower_histograms(sent_avg_arrays,title=title,labels=labels,colors=colors,num_bins=50,figsize=(10, 6),alpha=0.5,saveFig=False)
```

```python
title='Jensen Shannon Divergence Heatmap for Words'
labels = ['6v Delay', '6v Speak', '44 Delay', '44 Speak']
figPath='new_figs/jsd_words'
plot_js_heatmap_lower_triangle(chan_avg_arrays, title, labels, saveFig=False)
```

```python
labels = ['6v Delay', '6v Speak', '44 Delay', '44 Speak']
title = 'Jensen Shannon Divergence Heatmap for Sentences'
figPath='new_figs/jsd_sentences'
plot_js_heatmap_lower_triangle(sent_avg_arrays, title, labels, saveFig=False)
```
